```python
# Importando as bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Plotando um grafico de barra e colocando as legendas de X e Y
plt.bar(df_rh['Department'], df_rh['EmployeeNumber'])
plt.xlabel('Setor')
plt.ylabel('Funcionários')
```

```python
# Crie o gráfico de linha
Education = df_rh.groupby('Education')['EmployeeNumber'].count()

Education.plot(kind='line', marker='o')
plt.xlabel('ESCOLARIDADE')
plt.ylabel('QUANTIDADE DE FUNCIONARIOS ')
display(Education)
```

```python
# Criando grafico de pizza
atrito= df_rh.groupby('Attrition')[['EmployeeNumber']].sum()
explode = [0.1,0]
atrito.plot(kind = 'pie', subplots = True,autopct='%1.1f%%',explode=explode)

```
